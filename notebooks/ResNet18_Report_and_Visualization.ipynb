# ResNet-18 Report and Visualization

This notebook reads existing ResNet-18 outputs only (no retraining) and generates report-ready figures.

## 1) Load files

```python
from pathlib import Path
import pandas as pd
from IPython.display import Image, display

from src.report.generate_resnet_report_assets import main as generate_assets_main
```

```python
# Kaggle-friendly default paths
data_dir = Path('/kaggle/input/cifake-real-and-ai-generated-synthetic-images/CIFAKE')
resnet_dir = Path('outputs/resnet18')
output_dir = Path('outputs/resnet18_report_figures')

print('resnet_dir exists:', resnet_dir.exists())
print('output_dir:', output_dir)
```

## 2) Inspect results

```python
clean_df = pd.read_csv(resnet_dir / 'resnet18_clean_results.csv')
robust_df = pd.read_csv(resnet_dir / 'resnet18_robustness_results.csv')
train_df = pd.read_csv(resnet_dir / 'training_log.csv')

clean_df.head(), robust_df.head(), train_df.head()
```

## 3) Generate summary figures

```python
import sys
from src.report.generate_resnet_report_assets import main

sys.argv = [
    'generate_resnet_report_assets',
    '--data_dir', str(data_dir),
    '--resnet_clean_csv', str(resnet_dir / 'resnet18_clean_results.csv'),
    '--resnet_robustness_csv', str(resnet_dir / 'resnet18_robustness_results.csv'),
    '--training_log_csv', str(resnet_dir / 'training_log.csv'),
    '--confusion_matrix_png', str(resnet_dir / 'resnet18_confusion_matrix.png'),
    '--output_dir', str(output_dir),
]
main()
```

## 4) Save report assets
All generated assets are saved under `outputs/resnet18_report_figures/`.

## 5) Display generated figures

```python
figure_names = [
    'resnet18_clean_summary.png',
    'resnet18_training_curves.png',
    'resnet18_robustness_curves.png',
    'resnet18_avg_drop_bar.png',
    'resnet18_avg_drop_heatmap.png',
    'resnet18_confusion_matrix_copy.png',
    'corruption_examples.png',
]
for name in figure_names:
    p = output_dir / name
    if p.exists():
        display(Image(filename=str(p)))
    else:
        print('Missing:', p)
```
